```python
import csv
import time
import os 
```

```python
class EmployeeManager:
    def __init__(self):
        self.__Employees = {}
        self.filename = r"D:\Courses\Amit AI\Py\Final_project\Employees.csv"
        if os.path.exists(self.filename):
            f=open(self.filename,'r',newline='', encoding='utf-8')
            reader=csv.reader(f)
            next(reader)
            for i in reader:
                self.__Employees[int(i[0])] = {
                        "Name": i[1],
                        "Position": i[2],
                        "Salary": i[3],
                        "Email": i[4]
                    }
            f.close()
            
        if not os.path.exists(self.filename):
            f = open(self.filename, 'w', newline='', encoding='utf-8')
            writer = csv.writer(f)
            writer.writerow(["Id", "Name", "Position", "Salary", "Email"])
            f.close()
    def __save_updates(self):
        f = open(self.filename, 'w', newline='', encoding='utf-8')
        writer = csv.writer(f)
        writer.writerow(["Id", "Name", "Position", "Salary", "Email"])
        for num,data in self.__Employees.items():
            writer.writerow([num,data["Name"],data["Position"],data["Salary"],data["Email"]])
        f.close()

    def Add_Employees(self, ID, Name, Position, Salary, Email):
        self.__Employees[int(ID)] = {
            "Name": Name,
            "Position": Position,
            "Salary": Salary,
            "Email": Email
        }
        self.__save_updates()
        print("Employee added successfully.")
    
    def View_All_Employees(self):
        if self.__Employees :
            for num ,data in self.__Employees.items():
                print(num, data)
        else:
            print("No employees found.")
            
    def Update_Employee(self,ID, Name="", Position="", Salary="", Email=""):
        if int(ID) in self.__Employees:
            emp = self.__Employees[int(ID)]
            emp["Name"] = Name if Name else emp["Name"]
            emp["Position"] = Position if Position else emp["Position"]
            emp["Salary"] = Salary if Salary else emp["Salary"]
            emp["Email"] = Email if Email else emp["Email"]
            self.__save_updates()
        else:
            print("Employee not found.")
        
        
    def Delete_Employee(self,ID):
        if int(ID) in self.__Employees:
            del self.__Employees[int(ID)]
            self.__save_updates()
        else:
            print("Employee not found.")
    def Search_Employee(self,ID):
        if int(ID) in self.__Employees:
            print("Employee found:", self.__Employees[int(ID)])
        else:
            print("Employee not found.")
            
    def Exit(self):
        print("Exiting program. Goodbye!")
        return
```

```python
E1 = EmployeeManager()
while True:
    try:
            choice = int(input("\nEnter the number of operation you need: \n"
                               "1. Add Employee\n"
                               "2. Update Employee\n"
                               "3. Delete Employee\n"
                               "4. Search Employee\n"
                               "5. View All Employees\n"
                               "6. Exit\n"
                               "Choice: "))
        
            if choice == 1:
                no_of_emp=int(input("Enter number of Employees you need to add them"))
                for i in range (no_of_emp):
                    id=input("Enter an Employee ID :")
                    name=str(input("Enter an Employee name :"))
                    pos=str(input("Enter an Employee Position :"))
                    salary=input("Enter an Employee Salary :")
                    email=str(input("Enter an Employee Email :"))
                    if id =="" or name == "" or pos == "" or salary == "" or email == "":
                        print("All fields are required")
                    else:
                        try:
                            ID=int(id)
                            Salary=int(salary)
                            E1.Add_Employees(id,name,pos,salary,email)
                        except ValueError:
                            print("ID and Salary must be numbers")
            elif choice == 2:
                try:
                    id=int(input("Enter an Employee ID :"))
                    name=str(input("Enter an Employee name :"))
                    pos=str(input("Enter an Employee Position :"))
                    salary=input("Enter an Employee Salary :")
                    email=str(input("Enter an Employee Email :"))
                    E1 = EmployeeManager()
                    E1.Update_Employee(id,name,pos,salary,email)
                    print("Employee Updated successfully")
                    
                except ValueError:
                     print("ID and Salary must be numbers only")
            elif choice == 3:
                try:
                    id=int(input("Enter an Employee ID :"))
                    E1.Delete_Employee(id)
                    print("Employee deleted successfully")
                    
                except ValueError:
                     print("ID must be number only")
            elif choice == 4:
                try:
                    id=int(input("Enter an Employee ID :"))
                    E1.Search_Employee(id)
                except ValueError:
                     print("ID must be number only")
            elif choice == 5:
                E1.View_All_Employees()
            elif choice == 6:
                E1.Exit()
                break   
                
                
    except ValueError:
        print("Invalid input \nEnter number of operation only [1,2,3,4,5,6]") 
```
